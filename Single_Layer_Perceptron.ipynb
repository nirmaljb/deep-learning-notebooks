```python
import numpy as np
```

```python
x = np.array([1.0, 2.0])
y = 1
w = np.array([0.2, -0.5])
b = 0.1
eta = 0.1
```

```python
for epoch in range(4):
  print(f"Epoch : {epoch}")
  net_input = np.dot(x, w) + b
  print(f"net_input : {net_input}")
  y_pred = 1 if net_input > 0 else 0
  print(f"y_pred : {y_pred}")
  error = y - y_pred
  print(f"error: {error}")

  w = w + (eta * error) * x
  b = b + (eta * error)
  print(f"w : {w}")
  print(f"b : {b}")
```

```python
x_test = np.array([2.0,1.0])
net_test = np.dot(x_test, w) + b
y_pred = 1 if net_test > 0 else 0

#print statements
print(f"x_test : {x_test}")
print(f"net_test : {net_test}")
print(f"y_pred: {y_pred}")
```
